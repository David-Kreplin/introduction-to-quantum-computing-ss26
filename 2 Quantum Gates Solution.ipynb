### Calculating Statevectors and Quantum States from Quantum Circuits



Qiskit can calculate the resulting quantum state of a quantum circuit.
Reminder: the quantum state of a qubit can be written in two equivalent notations ($\alpha, \beta$ are amplitudes):

- Statevector: $\ket{\psi} = \begin{pmatrix} \alpha \\ \beta \end{pmatrix}$
- Quantum state (Dirac notation): $\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$

To determine both expressions for a given quantum circuit, we use the following functions from the helper library ``qc_lecture_tools``:

> - ``sv_vector(quantum_circuit, label=None)``: displays the resulting **statevector** of a quantum circuit
> 
> - ``sv_state(quantum_circuit, label=None)``: displays the resulting **quantum state** of a quantum circuit
> 
> - ``sv_probs(quantum_circuit, label=None)``: displays the resulting **measurement probabilities** after applying the quantum circuit

All functions have an optional second argument for a label. The representations can be displayed at any time using ``display()``.

```python
from qc_lecture_tools.statevector import sv_vector, sv_state, sv_probs  # Import helper functions
```

```python
# You can copy this code snippet throughout the lab to obtain the statevector,
# quantum state, and measurement probabilities of your quantum circuit.

# Example: Hadamard gate
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)

# Output of the statevector, quantum state, and measurement probabilities
display(sv_vector(quantum_circuit, "Statevector: "))
display(sv_state(quantum_circuit, "Quantum state: "))
display(sv_probs(quantum_circuit, "Measurement probabilities: "))
```

## Exercise 3: Multiple Gates

Next, we analyze how applying multiple gates in sequence affects the result and how this can be used to control the measurement probabilities.


**Exercise 3.1:** Analyze the effect of applying the following pairs of gates in succession: $HH$, $XX$, $YY$, $ZZ$. Investigate the effect either on the two basis states or, alternatively, on the 50/50 superposition state produced by the Hadamard gate (``.h()``).

```python
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
display(sv_state(quantum_circuit, "State before: "))
quantum_circuit.h(0)
quantum_circuit.h(0)
display(sv_state(quantum_circuit, "State after HH: "))
quantum_circuit.x(0)
quantum_circuit.x(0)
display(sv_state(quantum_circuit, "State after XX: "))
quantum_circuit.y(0)
quantum_circuit.y(0)
display(sv_state(quantum_circuit, "State after YY: "))
quantum_circuit.z(0)
quantum_circuit.z(0)
display(sv_state(quantum_circuit, "State after ZZ: "))
```

**Observation:** The gates $X$, $Y$, $Z$, and $H$ cancel when applied twice in succession. Therefore, a quantum circuit can sometimes be simplified.

**Exercise 3.2:** What happens when you apply two $S$ gates and two $T$ gates in succession? (Use the same procedure as in Exercise 3.1.)


```python
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
display(sv_state(quantum_circuit, "State before: "))
quantum_circuit.s(0)
quantum_circuit.s(0)
display(sv_state(quantum_circuit, "State after SS: "))
```

**Observation:** The result is the same as applying a $Z$ gate, so $SS = Z$.

```python
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
display(sv_state(quantum_circuit, "State before: "))
quantum_circuit.t(0)
quantum_circuit.t(0)
display(sv_state(quantum_circuit, "State after TT: "))
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
quantum_circuit.s(0)
display(sv_state(quantum_circuit, "State after S: "))
```

**Observation:** The result is the same as applying an $S$ gate, so $TT = S$.

**Exercise 3.4:** Investigate what happens when you apply the following gate combinations: $HXH$ and $HZH$.

```python
print("HXH on |0>:")
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
quantum_circuit.x(0)
quantum_circuit.h(0)
display(sv_vector(quantum_circuit, "Statevector: "))
display(sv_state(quantum_circuit, "Quantum state: "))
display(sv_probs(quantum_circuit, "Measurement probabilities: "))

quantum_circuit = QuantumCircuit(1)
print("HXH on |1>:")
quantum_circuit.x(0)
quantum_circuit.h(0)
quantum_circuit.x(0)
quantum_circuit.h(0)
display(sv_vector(quantum_circuit, "Statevector: "))
display(sv_state(quantum_circuit, "Quantum state: "))
display(sv_probs(quantum_circuit, "Measurement probabilities: "))
```

**Observation:** $HXH$ behaves like the $Z$ gate, so $HXH = Z$.

```python
print("HZH on |0>:")
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
quantum_circuit.z(0)
quantum_circuit.h(0)
display(sv_vector(quantum_circuit, "Statevector: "))
display(sv_state(quantum_circuit, "Quantum state: "))
display(sv_probs(quantum_circuit, "Measurement probabilities: "))

quantum_circuit = QuantumCircuit(1)
print("HZH on |1>:")
quantum_circuit.x(0)
quantum_circuit.h(0)
quantum_circuit.z(0)
quantum_circuit.h(0)
display(sv_vector(quantum_circuit, "Statevector: "))
display(sv_state(quantum_circuit, "Quantum state: "))
display(sv_probs(quantum_circuit, "Measurement probabilities: "))
```

**Observation:** $HZH$ behaves like the $X$ gate, so $HZH = X$.

**Exercise 3.5:** So far, we have only obtained probabilities of $p_0 = 1$, $p_1 = 1$, or $p_0 = 0.5$ and $p_1 = 0.5$.
Try using more than two gates from the set $\{X, Y, Z, H, S, T\}$ to obtain other probabilities. What do you notice?

```python
# No matter which gates are used here, the result remains 50% for |0> and 50% for |1>.
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
quantum_circuit.y(0)
quantum_circuit.z(0)
quantum_circuit.x(0)
quantum_circuit.s(0)
quantum_circuit.h(0)
display(sv_probs(quantum_circuit, "Measurement probabilities: "))
```

```python
# However, if a T gate is inserted, other probabilities can also occur.
quantum_circuit = QuantumCircuit(1)
quantum_circuit.h(0)
quantum_circuit.t(0)
quantum_circuit.y(0)
quantum_circuit.z(0)
quantum_circuit.x(0)
quantum_circuit.s(0)
quantum_circuit.h(0)
display(sv_probs(quantum_circuit, "Measurement probabilities: "))
```

**Observation:** $T$ gates are required to achieve probabilities beyond $0$,$0.5$ and $1$.
